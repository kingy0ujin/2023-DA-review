### 기온 데이터를 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] !='':
        result.append(float(row[-1]))

plt.hist(result, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []


for row in data:
     if len(row) > 0 and '-' in row[0]:
        month = row[0].split('-')[1]
        if row[-1] != '' :
            if month == '08':
                aug.append(float(row[-1]))

plt.hist(aug, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
     if len(row) > 0 and '-' in row[0]:
        month = row[0].split('-')[1]
        if row[-1] != '':
            if month == '08':
                aug.append(float(row[-1]))
            if month =='01':
                jan.append(float(row[-1]))

plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```

### 기온 데이터를 상자 그림으로 표현하기

```python
import matplotlib.pyplot as plt
import random
result = []

for i in range(13):
    result.append(random.randint(1, 1000))
print(sorted(result))

plt.boxplot(result)
plt.show()
```

```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    #자꾸 인덱스 오류가 나서 추가한 조건문
    if len(row) > 0 and '-' in row[0]:
        month = row[0].split('-')[1]
        if row[-1] != '':
            if month == '08':
                aug.append(float(row[-1]))
            if month =='01':
                jan.append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False)

plt.show()
```
