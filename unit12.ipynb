### 출근 시간대 사람들이 가장 많이 타고 내리는 역 찾기

### 밤 11시에 사람들이 가장 많이 타고 내리는 역 찾는 코드

```python
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

next(data)
next(data)

mx = 0
mx_station = ''
t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? : '))

for row in data:
    for i in range(4,52):
        row[i] = int(row[i].replace(",","")) # map(int, row[4:]) 이지만 오류떠서 바꿈
    a = row[4+(t-4)*2] # 특정 시간대의 승차 인원
    if a > mx:
        mx = a
        mx_station = row[3] + '(' + row[1] + ')'
print(mx_station, mx)
```

### 시간대별 사람들이 가장 많이 타고 내리는 역은 어디일까

```python
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

next(data)
next(data)
mx = [0] *24
mx_station = [''] *24

for row in data:
    for i in range(4,52):
        row[i] = int(row[i].replace(",","")) # map(int, row[4:]) 이지만 오류떠서 바꿈
    for j in range(24):
        a = row[j*2+4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]
print(mx_station)
print(mx)

```

```python
import csv
import matplotlib.pyplot as plt
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

next(data)
next(data)
mx = [0] *24
mx_station = [''] *24

for row in data:
    for i in range(4,52):
        row[i] = int(row[i].replace(",","")) # map(int, row[4:]) 이지만 오류떠서 바꿈
    for j in range(24):
        a = row[j*2+4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]+ '(' + str(j+4) + ')'
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

### 시간대별 하차 인원이 가장 많은 역을 찾는 코드

```python
import csv
import matplotlib.pyplot as plt
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)

next(data)
next(data)
mx = [0] *24
mx_station = [''] *24

for row in data:
    for i in range(4,52):
        row[i] = int(row[i].replace(",","")) # map(int, row[4:]) 이지만 오류떠서 바꿈
    for j in range(24):
        b = row[j*2+5]
        if b > mx[j]:
            mx[j] = b
            mx_station[j] = row[3]+ '(' + str(j+4) + ')'
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

### 지하철 시간대별 승하차 인원 추이를 나타내는 코드

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

s_in = [0] *24
s_out = [0] *24

for row in data:
    for i in range(4,52):
        row[i] = int(row[i].replace(",","")) # map(int, row[4:]) 이지만 오류떠서 바꿈
    for j in range(24):
        s_out[j] += row[j*2+5]
        s_in[j] += row[j*2+4]

plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24),range(4, 28))
plt.show()
```
